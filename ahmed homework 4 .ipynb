```python
import random
import matplotlib.pyplot as plt

def monte_carlo_simulation(num_trials):
    stick_wins = 0
    switch_wins = 0

    for _ in range(num_trials):
        # Initialize the boxes
        boxes = ['10', '10', '100']

        # Player selects a box
        player_choice = random.choice([0, 1, 2])

        # Friend opens another box with $10
        friend_choice = random.choice([i for i in range(3) if i != player_choice and boxes[i] == '10'])

        # Check if the player sticks with their choice or switches
        if boxes[player_choice] == '100':
            stick_wins += 1
        else:
            switch_wins += 1

    return stick_wins, switch_wins

# Set the number of Monte Carlo trials
num_trials = 100000

# Perform the Monte Carlo simulation
stick_wins, switch_wins = monte_carlo_simulation(num_trials)

# Calculate probabilities
stick_probability = stick_wins / num_trials
switch_probability = switch_wins / num_trials

# Create a bar plot to visualize the probabilities
labels = ['Stick', 'Switch']
probabilities = [stick_probability, switch_probability]

plt.bar(labels, probabilities)
plt.xlabel('Choice')
plt.ylabel('Probability')
plt.title('Probability of Winning: Stick vs. Switch')
plt.ylim(0, 1)
plt.show()

print("Probability of winning if you stick:", stick_probability)
print("Probability of winning if you switch:", switch_probability)
```

```python
import random
import sympy as sp

def f(x):
    return x * sp.cos(71 * x) + sp.sin(13 * x)

x = sp.Symbol('x')
expression = f(x)

def monte_carlo_integration(f, a, b, num_samples):
    total = 0

    for _ in range(num_samples):
        x_val = random.uniform(a, b)
        total += f.subs(x, x_val)

    average = total / num_samples
    integral = (b - a) * average
    return integral

# Set the number of samples for the Monte Carlo simulation
num_samples = 100000

# Perform the Monte Carlo integration
integral_estimate = monte_carlo_integration(expression, 0, 1, num_samples)

print("Estimated Integral:", integral_estimate)
```
